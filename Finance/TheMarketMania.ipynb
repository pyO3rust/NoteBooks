```python
#To count the numbers using for loop

def count(numbers):
    total = 0
    for x in numbers:
        if x < 40:
            total += 1  #incrementing from 0 //just like result  //total is the position
    return total

list = [12, 34, 11, 56, 31, 34, 12, 50]
count(list)
```

```python
#to count the numbers using while loop

def count(numbers):
    numbers = sorted(numbers) #without sorting it will go/exit out of loop from 56 
    total = 0
    while numbers[total] < 40: #assignment[key] = value
        total = total + 1
    return total
list = [12, 34, 11, 56, 31, 34, 12, 50]
count(list)   
```

# Iterating Over Dictionaries

```python
#money spent on products with 5 dollars ormore than 5 dollars
prices = {
    "Classmate" : 35,
    "Pencil" : 4,
    "Colors" : 25,
    "Eraser" : 3
}
quantity = {
    "Classmate" : 5,
    "Pencil" : 10,
    "Colors" : 2,
    "Eraser" : 3,
}

money_spent = 0

for i in prices:
    if prices[i] >= 5:
        money_spent += (prices[i]*quantity[i])
    else:
        money_spent = money_spent
    
print(money_spent)
```

# Simple Rate of Return


$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

$${P_1} = ending price$$  $${P_0} = beginning price$$

```python
#happiestMinds['simple_return'] = creating a column, shift(1) = pulls out the previous day data
#pandas.DataFrame.shift(# of lags) in this case 1
#daily rate of returns
happiestMinds['simple_return'] = (happiestMinds['Adj Close']/happiestMinds['Adj Close'].shift(1)) - 1
happiestMinds
```

```python
print(happiestMinds['simple_return'])
```

```python
happiestMinds.tail(16)
```

```python
#to create a graph of rate of return
plot = happiestMinds['simple_return'].plot(figsize=(8, 5)) #figuresize = size of the graph, 8 = width & 5 = height
plt.show(plot)
```

or

```python
happiestMinds['simple_return'].plot(figsize=(8, 5))
plt.show()
```

```python
#average daily rate of returns 
avg_returns_daily = happiestMinds['simple_return'].mean()
avg_returns_daily
```

```python
#avg returns per year
avg_returns_yearly = happiestMinds['simple_return'].mean() * 250 #250 = trading days per year
avg_returns_yearly
```

```python
#rounding off to 5digits & converting yearly returns into percentage
percent = str(round(avg_returns_yearly, 5) * 100) + '%'
print(percent)
```

# Calculating the Return of a Portpolio of Securities

```python
returns = (newDataUp / newDataUp.shift(1)) - 1
returns.tail()
```

```python
show_graph = returns.plot(figsize=(16, 8)) #figuresize = size of the graph, 8 = width & 5 = height
plt.show(show_graph)
```

```python
#average rate of return daily
average_daily = returns.mean()
average_daily
```

```python
#assigning the weights(sum of array should be equal to 1)
weights = np.array([0.25, 0.25, 0.25, 0.25, 0.25])
```

```python
#calculating the dot product
dotProd = np.dot(average_daily, weights)
print(dotProd)
```

```python
#daily rate of return percentage of the portfolio
myPortf = str(round(dotProd, 5) * 100) + '%'
print(myPortf)
```

```python
#annual rate of returns % of the portfolio on the trading days using daily rate
annually = str(0.126 * 250) + '%'
print(annually)
```

```python
#average annual rate of returns returns using mean
average_annual = returns.mean() * 250
average_annual
```

```python
#calculating the dot product(1column matrix dot product 1 column matrix is a single output, weight already assigned up )
np.dot(average_annual, weights)
```

```python
#calculating the rate of return of a portfolio annually
portf_1 = str(round(np.dot(average_annual, weights), 5) * 100) + '%'
print(portf_1)
```

```python
#finding the annual returns % for diff weightage of portfolio
weights_1 = np.array([0.60, 0.15, 0.10, 0.05, 0.10])
pFolio = str(round(np.dot(average_annual, weights_1) , 5) * 100) + '%'
print(pFolio)
```

# Calculating the beta of the stock

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb

tickers = ['WIPRO.NS', '^NSEI']
data = pd.DataFrame()
for t in tickers:
    data[t] = wb.DataReader(t, data_source='yahoo', start='2016-1-1', end='2021-1-1')['Adj Close'] 

data
```

```python
sec_returns = np.log(data / data.shift(1))
```

```python
cov = sec_returns.cov() * 250
cov
```

```python
#obtaining covariance of the market as the float
cov_with_market = cov.iloc[0,1]
cov_with_market 
```

```python
#annualized variance of nifty 50 as the market
market_var = sec_returns['^NSEI'].var() * 250 
market_var
```

**BETA**
### $$ 
\beta_{pg} = \frac{\sigma_{pg,m}}{\sigma_{m}^2}
$$

```python
wipro_beta = cov_with_market / market_var
wipro_beta
```

**Calculate the expected return of P&G (CAPM):**
### $$
\overline{r_{pg}} = r_f + \beta_{pg}(\overline{r_{m}} - r_f) 
$$

```python
#this is the expected return a person would want buying wipro stocks i.e nearly 10 percent
#er for expected returns, 0.062 is the 10 yrs bond yield & 0.065 is the equity risk premium
wipro_er = 0.062 + wipro_beta * 0.065
wipro_er
```

### William Sharpe's Sharpe Ratio

**Sharpe ratio:**
### $$
Sharpe = \frac{\overline{r_{pg}} - r_f}{\sigma_{pg}}
$$

```python
sharpe = (wipro_er - 0.062) / (sec_returns['WIPRO.NS'].std() * 250 ** 0.5)
sharpe
```
